```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import rotate
```

```python
img = cv2.imread('pexel1.jpg')
img2 = cv2.imread('download.jpeg')

# img = img[:190,:265]

  
```

```python
img.shape,img2.shape
```

```python
def imshow(im):
    cv2.imshow('Preview',im)
    cv2.waitKey(0)  #wait infinitely untill any key is pressed
    cv2.destroyAllWindows()
```

```python
def showimg(im):
    plt.imshow(im[:,:,::-1])
    plt.axis()
    plt.show()
def imshow(im):
    cv2.imshow('Preview',im)
    cv2.waitKey(0)  #wait infinitely untill any key is pressed
    cv2.destroyAllWindows()
    
```

```python
imshow(img)
imshow(img2)
```

```python
added = cv2.add(img, img2)
showimg(added)
   
```

```python
subtracted = cv2.subtract(img, img2)
showimg(subtracted)
```

```python
img2_1 = cv2.multiply(img, 2 *np.ones(img2.shape).astype('uint8'))
showimg(img2_1)
```

```python
image_rotated = rotate(img, angle=45, resize=True)
showimg(image_rotated)
```

```python
from time import sleep
repeat = True
while repeat:
    for i in range(0,361,5):
        i = rotate(img, angle= i, resize = True)
        cv2.imshow('',i)
        key = cv2.waitKey(1)
        if key==ord('q'):
            repeat = False
            break
        sleep(0.1)
cv2.destroyAllWindows()
```

-->image morphology
dilation- to grow,
erosion- to remove,
opening- remove then grow,first erode then dilate.  
closing- grow then remove
